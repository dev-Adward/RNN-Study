```python
# Generate the sample data
model_year = np.random.randint(2000, 2023, 100)
mileage = np.random.randint(1000, 200000, 100)
price = 30000 - 0.1 * model_year + 0.05 * mileage + np.random.randn(100) * 5000

X = np.column_stack((model_year, mileage))

# Scale the features
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X = X.reshape(X.shape[0], 1, X.shape[1])

# Scale the target variable
scaler_price = MinMaxScaler()
y = scaler_price.fit_transform(price.reshape(-1, 1))

# Create the LSTM model
model = Sequential()
model.add(LSTM(units=256, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(100, activation='tanh'))
model.add(Dense(1))  # For regression, so no activation function

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=300, batch_size=32)
```

```python
train_predictions = model.predict(X[:10])
train_predictions = scaler_price.inverse_transform(train_predictions)  # Reverse scaling

# Evaluate the model on the test data
test_predictions = model.predict(X[-10:])
test_predictions = scaler_price.inverse_transform(test_predictions)  # Reverse scaling

# Calculate Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error

train_mse = mean_squared_error(y[:10], train_predictions)
test_mse = mean_squared_error(y[-10:], test_predictions)
print('Mean Squared Error (Train):', train_mse)
print('Mean Squared Error (Test):', test_mse)

# Visualize the predictions
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(y[:10], train_predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('Training Set Predictions')

plt.subplot(1, 2, 2)
plt.scatter(y[-10:], test_predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('Test Set Predictions')

plt.tight_layout()
plt.show()
```
